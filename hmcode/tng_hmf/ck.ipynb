```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
data = np.loadtxt('D:\WHS\code\HMcode-python/tng_hmf\BoverD.txt', dtype=str, skiprows=1)
arrays = list(zip(*data))
x = np.array(arrays[0], dtype=float)
y = np.array(arrays[1], dtype=float)

datas = np.loadtxt("D:\WHS\code\HMcode-python/tng_hmf/nu.txt", skiprows=1)
nu = np.array(datas, dtype=float)

print(x)
```

```python
plt.plot(nu, y)
```

```python
intp = interpolate.interp1d(nu, y, kind='linear')

nunew = np.linspace(nu.min(), nu.max(), 100)
ynew = intp(nunew)


plt.plot(nunew, ynew, 'r')
plt.show()
```

```python
nue = 1
ynue = intp(nue)
print(ynue)
```
