```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from kllr import *
from sklearn.neighbors import LocalOutlierFactor
```

```python
z = np.arange(0, 4.5, 0.5)
```

```python
hmf = np.loadtxt(f'/data/wbc/CAMELS/alldataprocessing/hmf/data/supression_0.0.txt')
```

```python
mask = np.isnan(hmf[:,1]) | np.isinf(hmf[:,1])
hmf = hmf[~mask]
plt.plot(hmf[:,0],hmf[:,1])
y = hmf[:,1]
mmm = hmf[:,0]
```

```python
X = y.reshape(-1, 1)

# 计算 LOF
lof = LocalOutlierFactor(n_neighbors=30)
y_pred = lof.fit_predict(X)  # 1: 正常点, -1: 异常点
lof_scores = lof.negative_outlier_factor_  # LOF 得分

# 识别离群点
outliers = X[y_pred == -1]

# 可视化结果
plt.figure(figsize=(8, 4))
plt.scatter(X, np.zeros_like(X), c=y_pred, cmap='coolwarm', edgecolors='k', s=80)
plt.scatter(outliers, np.zeros_like(outliers), c='red', edgecolors='k', marker='x', s=100, label="Outliers")
plt.axvline(np.percentile(lof_scores, 5), linestyle="--", color="gray", label="Threshold")
plt.title("LOF Outlier Detection")
plt.legend()
plt.show()

# 输出异常点
print("检测到的离群点:", outliers.flatten())
```

```python
outxs = mmm[y_pred == -1]
plt.scatter(mmm, y, s=1)
plt.scatter(outxs, outliers.flatten(), color='r', label='outliers', s=1, marker='x')
plt.plot(mmm[y_pred == 1], y[y_pred == 1], color='k', label='inliers', linewidth=0.5)
plt.xlim(10, )
#plt.ylim(0.8,1.5)
```

```python
for i in range(len(z)):
    hmf = np.loadtxt(f'/data/wbc/CAMELS/alldataprocessing/hmf/data/supression_{z[i]}.txt')
    mask = np.isnan(hmf[:,1]) | np.isinf(hmf[:,1])
    hmf = hmf[~mask]
    plt.plot(hmf[:,0],hmf[:,1])
    plt.show()

    X = hmf[:,1].reshape(-1, 1)

    # 计算 LOF
    lof = LocalOutlierFactor(n_neighbors=40)
    y_pred = lof.fit_predict(X)  # 1: 正常点, -1: 异常点
    lof_scores = lof.negative_outlier_factor_  # LOF 得分

    # 识别离群点
    outliers = X[y_pred == -1]

    # 可视化结果
    plt.figure(figsize=(8, 4))
    plt.scatter(X, np.zeros_like(X), c=y_pred, cmap='coolwarm', edgecolors='k', s=80)
    plt.scatter(outliers, np.zeros_like(outliers), c='red', edgecolors='k', marker='x', s=100, label="Outliers")
    plt.axvline(np.percentile(lof_scores, 5), linestyle="--", color="gray", label="Threshold")
    plt.title("LOF Outlier Detection")
    plt.legend()
    plt.show()

    hmf = hmf[y_pred == 1, :]

    plt.plot(hmf[:,0],hmf[:,1])
    plt.show()

    df = pd.DataFrame(hmf, columns = ['m', 'supression'])
    # Setup kernel local linear regression model
    data, ax = Plot_Fit_Summary(df, 'm', 'supression', show_data=True, kernel_width = 0.2)

    if np.max(data['x']) < 14:
        np.concatenate([data['x'], np.array([14])])
        np.concatenate([data['y'], np.array([1])])
    np.savetxt(f'/data/wbc/CAMELS/HMcode-python/hmcode/hmf/regression/regression_{z[i]}.txt', np.column_stack((data['x'], data['y'], [z[i]]*len(data['x']))))
```
